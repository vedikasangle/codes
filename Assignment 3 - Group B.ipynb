```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(r"C:\Users\Vedika\Downloads\airquality1.csv")
```

```python
data
```

```python
data["Ozone"] = data["Ozone"].fillna(data["Ozone"].mean())
data["Solar.R"] = data["Solar.R"].fillna(data["Solar.R"].mean())
```

```python
data
```

```python
data.isnull()
```

```python
sizes = [data["Ozone"].mean(),data["Solar.R"].mean(),data["Temp"].mean(),data["Wind"].mean()]
labels = 'Ozone', 'Solar.R' , 'Temp', 'Wind'
plt.pie(sizes, autopct="%1.2f%%", labels = labels)
plt.plot()
```

```python
h = data.iloc[1:21,2]
```

```python
y_pos = np.arange(len(h))
v= range(1,21)
```

```python
plt.bar(y_pos,h)
plt.xticks(y_pos,v)
plt.xlabel('Days')
plt.ylabel('Range')
plt.title('Temperature for 1st - 20th May' )
plt.show()
```
