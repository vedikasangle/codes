### Generating Transaction List

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(A.values[i,j]) for j in range(0, 20)])
```

### Creating apriori model

```python
from apyori import apriori
tran_rules = apriori(transactions, min_support = 0.003, min_confidence = 0.6, 
                     min_lift = 3, min_length = 2)
```

### Visualising the results

```python
results = list(tran_rules)
print(results[:10])
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))


    
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 
                            'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame.nlargest(n = 10, columns = 'Confidence')
```
