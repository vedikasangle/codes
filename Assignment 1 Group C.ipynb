```python
import requests #used to request access
import bs4 #to pull data from website
```

```python
request1 = requests.get("https://www.flipkart.com/philips-hp8302-06-hair-straightener/p/itmf51277d7653cd?pid=HSTE4YPPY43YAXJZ&lid=LSTHSTE4YPPY43YAXJZNCZAJ9&marketplace=FLIPKART&store=zlw%2F79s%2Fdk5&spotlightTagId=FkPickId_zlw%2F79s%2Fdk5&srno=b_1_3&otracker=browse&fm=organic&iid=884b85b5-4fbe-4077-b2a1-c48a6bd72347.HSTE4YPPY43YAXJZ.SEARCH&ppt=browse&ppn=browse&ssid=i4gvu8c7og0000001684896358929")
```

```python
request1.content
```

```python
soup = bs4.BeautifulSoup(request1.text)
```

```python
print(soup)
```

```python
#fetching reviews
reviews = soup.find_all('div', {'class' : 't-ZTKy'});
for review in reviews:
    print(review.get_text() + "\n")
```

```python
#fetching rating
ratings = soup.find('div', {'class' : '_2d4LTz'}).get_text()
print(ratings)
```

```python
#fetching customer name
customer_name = soup.find_all('p', {'class':'_2sc7ZR _2V5EHH'});
for cust_name in customer_name:
    print(cust_name.get_text() + "\n")
```

```python
#fetching reviews(short)
shortreviews = soup.find_all('p', {'class' : '_2-N8zT'});
for review in shortreviews:
    print(review.get_text() + "\n")
```

```python
#fetching price
ratings = soup.find('div', {'class' : '_30jeq3 _16Jk6d'}).get_text()
print(ratings)
```
